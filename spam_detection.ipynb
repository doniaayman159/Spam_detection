````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.naive_bayes import BernoulliNB,ComplementNB,MultinomialNB,GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
````

````python
#make name columns

columun_names="word_freq_make,word_freq_address,word_freq_all,word_freq_3d,word_freq_our,word_freq_over,word_freq_remove,word_freq_internet,word_freq_order,word_freq_mail,word_freq_receive,word_freq_will,word_freq_people,word_freq_report,word_freq_addresses,word_freq_free,word_freq_business,word_freq_email,word_freq_you,word_freq_credit,word_freq_your,word_freq_font,word_freq_000,word_freq_money,word_freq_hp,word_freq_hpl,word_freq_george,word_freq_650,word_freq_lab,word_freq_labs,word_freq_telnet,word_freq_857,word_freq_data,word_freq_415,word_freq_85,word_freq_technology,word_freq_1999,word_freq_parts,word_freq_pm,word_freq_direct,word_freq_cs,word_freq_meeting,word_freq_original,word_freq_project,word_freq_re,word_freq_edu,word_freq_table,word_freq_conference,char_freq_;,char_freq_(,char_freq_[,char_freq_!,char_freq_$,char_freq_#,capital_run_length_average,capital_run_length_longest,capital_run_length_total,spam"
columun_names=columun_names.split(sep=',')
columun_names
````

Load **DataSet**

````python


#Load **DataSet**

df=pd.read_csv("./Spam_data.txt",header = None,sep=',',names=columun_names)

df

````

````python
df.spam
````

Function to Spilt DataSet by rate like (80,20)

````python

#Function to Spilt DataSet by rate like (80,20)

def split_data(data,rate):

  last_element=round(len(data)*rate)
  print(last_element)
  #for make suffle in data

  #data= data.reindex(np.random.permutation(data.index))

  #################
  X_train=data.drop(labels='spam',axis=1)[:last_element]
  y_train=data.spam[:last_element]
  X_test=data.drop(labels="spam",axis=1)[last_element:]
  y_test=data.spam[last_element:]
  return X_train, y_train, X_test,y_test
````

###################################################

````python
X_first_80_percent_train, Y_first_80_percent_train, X_last_20_percent_test,Y_last_20_percent_test=split_data(df,.8)
````

# Model applying

````python
# Model applying



model_GNB = GaussianNB()
model_MNG = MultinomialNB()

# fit model
model_GNB.fit(X_first_80_percent_train, Y_first_80_percent_train)
model_MNG.fit(X_first_80_percent_train, Y_first_80_percent_train)

#test model
predicted_GNB = model_GNB.predict(X_last_20_percent_test)
predicted_MNG = model_MNG.predict(X_last_20_percent_test)

````

calculate the accuracy

````python


# calculate the accuracy
accuracy_GNB=accuracy_score(Y_last_20_percent_test,predicted_GNB)
accuracy_MNG=accuracy_score(Y_last_20_percent_test,predicted_MNG)


print("accuracy_GNB:", accuracy_GNB)
print("accuracy_MNG:", accuracy_MNG)
````

**Plot** the Confusion Matrix

````python


# Plot the Confusion Matrix for GaussianNB

ConfusionMatrixDisplay.from_estimator( model_GNB, X_last_20_percent_test, Y_last_20_percent_test)
plt.title("GaussianNB 20_percent_test data set")
plt.show()
````

````python
# Plot the Confusion Matrix for MultinomialNB

ConfusionMatrixDisplay.from_estimator( model_MNG, X_last_20_percent_test, Y_last_20_percent_test)
plt.title("model_MultinomialNB 20_percent_test data set")
plt.show()
````

````python
X=df.drop('spam',axis=1)
Y=df['spam']
````

````python

#split dataset  by useing the train_test_split function
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
````


# Model applying

````python

# Model applying


model_GNB = GaussianNB()
model_MNG = MultinomialNB()

#fit models

model_GNB.fit(X_train, Y_train)
model_MNG.fit(X_train, Y_train)

#test model

predicted_GNB = model_GNB.predict(X_test)
predicted_MNG = model_MNG.predict(X_test)
````

````python


# calculate the accuracy

accuracy_GNB=accuracy_score(Y_test,predicted_GNB)
accuracy_MNG=accuracy_score(Y_test,predicted_MNG)


print("accuracy_GNB:", accuracy_GNB)
print("accuracy_MNG:", accuracy_MNG)
````

````python

M = confusion_matrix(Y_test, predicted_GNB)

ConfusionMatrixDisplay(M )
````

````python
ConfusionMatrixDisplay.from_estimator( model_GNB, X_test, Y_test)
# Plot the Confusion Matrix for GaussianNB
plt.title("model_GaussianNB train_test_split data")
plt.show()
````

````python
# Plot the Confusion Matrix for MultinomialNB
ConfusionMatrixDisplay.from_estimator( model_MNG, X_test, Y_test)
plt.title("model_MultinomialNB train_test_split data")
plt.show()
````

````python


# applying new models

model_BNB = BernoulliNB()
model_CNB = ComplementNB()

# fit these model
model_BNB.fit(X_train, Y_train)
model_CNB.fit(X_train, Y_train)

# test these model
predicted_BNB = model_GNB.predict(X_test)
predicted_CNB = model_CNB.predict(X_test)
````

````python

#calcuate the accuracy

accuracy_BNB=accuracy_score(Y_test,predicted_BNB)
accuracy_CNB=accuracy_score(Y_test,predicted_CNB)


print("accuracy_GNB:", accuracy_BNB)
print("accuracy_CNB:", accuracy_CNB)
````

````python

# Plot the Confusion Matrix for BernoulliNB

ConfusionMatrixDisplay.from_estimator( model_BNB, X_test, Y_test)

plt.title("model_BernoulliNB train_test_split data")
plt.show()
````
